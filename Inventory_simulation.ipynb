```python
from SimulationSrc import *
import openai
from dotenv import load_dotenv

load_dotenv('.env')
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
results = []
# 共告知day天的数据
for day in range(1, 6):
    # 重复模拟20次
    for idx in range(1, 21):
        sl = InventorySimulator()
        # data = sl.simulate_constOrder(day)
        data = sl.simulate(day)

        print(f'{day} day {idx} times: {data}')
        DemandLambda = sl.get_DemandLambda()
        OrderLambda = sl.get_OrderLambda()
        # 只用Chat
        # 最后一行是结果，是单个数字
        chatGPTinfer = Inventory_GPTInfer(data)
        base_prompt = chatGPTinfer.get_prompt_inventory()
        rl_prompt = chatGPTinfer.get_prompt_rl()
        chatGPTresult = chatGPTinfer.infer()
        chatGPTresult_rl = chatGPTinfer.infer_rl()

        # 同时保存 data
        results.append([day, idx, data, DemandLambda, OrderLambda, base_prompt, chatGPTresult, rl_prompt, chatGPTresult_rl])
        print('finish', day, idx)
```

```python
import csv
with open('./results/result_0410_inventory.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['day', 'idx', 'SampleData', 'DemandLambda', 'OrderLambda' ,'base_prompt', 'chatGPTresult', 'rl_prompt', 'chatGPTresult_rl'])
    for row in results:
        writer.writerow(row)
```
